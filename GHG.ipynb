```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/SupplyChainEmissionFactorsforUSIndustriesCommodities2015_Summary.csv'
df = pd.read_csv(file_path)

# Clean column names
df.columns = df.columns.str.strip()

# Define correct column names
emission_col = 'Supply Chain Emission Factors with Margins'
industry_col = 'Industry Name'

# Sort and select top 10 industries by GHG emissions
top10 = df.sort_values(by=emission_col, ascending=False).head(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=emission_col, y=industry_col, data=top10, palette='Reds_r')
plt.title('Top 10 Industries by GHG Emissions (with Supply Chain Margins)')
plt.xlabel('GHG Emissions (kg CO2e per $)')
plt.ylabel('Industry')
plt.tight_layout()
plt.show()
```
